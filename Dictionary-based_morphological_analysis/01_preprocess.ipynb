```python
import os
import pandas as pd
import re
```

```python
df = pd.read_csv('/content/drive/MyDrive/menu_labelling/data/data.csv')
```

```python
df.head()
```

```python
cols = ['store','MENU_NM','표준화상품명']
```

```python
df = df[cols]
df = df.sort_values('MENU_NM', ascending = True)
```

#0. NA값 채우기

```python
df = df.fillna('missing')
```

# 1. 특수기호 제거

```python
df['MENU_NM2'] = df['MENU_NM'].str.replace(pat=r'/|·|_|♡|★|▶|★|☆|●|-|!|™|#', repl=r'', regex=True)
```

# 2. 괄호 제거

```python
df['MENU_NM3'] = df['MENU_NM2'].str.replace('[\[\]()]', '', regex=True)
```

#3. 'ㄷ'으로 시작하는 데이터 뽑아내기

```python
def filter_g_rows(df, column_name):
    return df[df[column_name].apply(lambda x: '다' <= x[0] <= '딯' if x and len(x) > 0 else False)]
```

```python
filtered = filter_g_rows(df, 'MENU_NM3')
```

#4. 저장

```python
filtered.to_csv('/content/drive/MyDrive/menu_labelling/\'ㄷ\'으로 시작하는 데이터/d_filtered_preprocess.csv',index=False)
```
