```python
import os
import pandas as pd
import jamo
from jamo import h2j, j2hcj
```

#2. 종성 함수 만들기

```python
def get_jongsung_TF(sample_text):
    sample_text_list = list(sample_text)
    last_word = sample_text_list[-1]
    last_word_jamo_list = list(j2hcj(h2j(last_word)))
    last_jamo = last_word_jamo_list[-1]

    jongsung_TF = "T"

    if last_jamo in ['ㅏ', 'ㅑ', 'ㅓ', 'ㅕ', 'ㅗ', 'ㅛ', 'ㅜ', 'ㅠ', 'ㅡ', 'ㅣ', 'ㅘ', 'ㅚ', 'ㅙ', 'ㅝ', 'ㅞ', 'ㅢ', 'ㅐ,ㅔ', 'ㅟ', 'ㅖ', 'ㅒ']:
        jongsung_TF = "F"

    return jongsung_TF
```

#3. 종성 함수 적용

```python
# 데이터 불러오기
dict_notf = pd.read_csv('/content/drive/MyDrive/menu_labelling/dictionaries/dict1_v1_0.csv')
dict_notf.head()
```

```python
dict_notf['<has_jongseong>'] = dict_notf['<surface>'].apply(get_jongsung_TF)
```
