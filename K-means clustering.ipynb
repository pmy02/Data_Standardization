```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from google.colab import drive

drive.mount('/content/drive')

# 데이터 로드
df = pd.read_excel("/content/drive/My Drive/상품정보 표준화/표준화 4.xlsx")
product_names = df["표준상품명"].tolist()
```

```python
# 텍스트 데이터를 벡터 형태로 변환
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(product_names)

# 클러스터 개수 설정 및 K-means 알고리즘 적용
num_clusters = 50
kmeans = KMeans(n_clusters=num_clusters, random_state=0).fit(X)

pd.set_option('display.max_rows', 1000)

# 클러스터링 결과 확인
df["클러스터"] = kmeans.labels_
print(df.head())
```

```python
# 클러스터 별 상품명 확인
for cluster in range(num_clusters):
    print(f"클러스터 {cluster}:")
    print(df[df["클러스터"] == cluster]["표준상품명"])
```

```python
# 클러스터별 대표 상품명 보기기
for cluster in range(num_clusters):
    products_in_cluster = df[df["클러스터"] == cluster]["표준상품명"].tolist()
    representative_product_name = max(set(products_in_cluster), key=products_in_cluster.count)
    print(f"클러스터 {cluster} 대표 상품명: {representative_product_name}")
```
